# PyCity Schools Analysis written description

 -	There are total of 39170 students and 15 schools in this district with a total budget of 82,932,329,558. 

Variables used measure and compare school performances include:

-	Average Math Score (AMS)
-	Average Reading Score (ARS)
-	Percentage Passing Math (PPM)
-	Percentage Passing Reading (PPR)
-	Overall Passing Rate (OPR)

District summary analysis showed that in this school district:
-	AMS = 78.985371
-	ARS = 81.87784
-	PPM =74.980853
-	PPR = 85.805463
-	OPR = 80.431606

Analysis variables by school type showed that Charter schools have higher performances than District schools. 
-	Charter schools: AMS = 83.473852, ARS = 83.896421, PPM = 93.620830, PPR=96.586489, OPR=95.103660
-	District schools: AMS = 76.956733, ARS = 80.966636, PPM = 66.548453, PPR=80.799062, OPR=73.673757


As evidenced by school summary analysis, five charter schools are found among top performing schools.  School summary analysis also showed that district schools are among the lower five bottom performing school.  

Analysis of AMS  and ARS  by grade (9th,10th, 11th ,12th) showed that there is no significance variation on scores across all grades in all schools. 

Analysis of variables by school spending showed that there is an inverse relationship between budget per student and variables scores. Increased spending doesn’t improve scores. 

Analysis of variables by school size showed that there is a direct relationship between school size and variables scores. As school size increases performance decreases. 

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
schooldata = "schools_complete.csv"
studentdata = "students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data_df = pd.read_csv(schooldata )
student_data_df = pd.read_csv(studentdata)

# Combine the data into a single dataset
school_data_complete= pd.merge(school_data_df, student_data_df, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

#  1.	PyCitySchools:-District Summary

##  all tasks are completed

-	Calculate the average math score

-	Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

-	Calculate the percentage of students with a passing reading score (70 or greater)
-	Create a dataframe to hold the above results

```python
Total_School = len(school_data_complete["school_name"].unique())
Total_School
```

```python
# Calculate the total number of students
Total_Students = school_data_complete["Student ID"].count()
Total_Students
```

```python
# Calculate the total budget
Total_Budget = school_data_complete["budget"].sum()
Total_Budget
```

```python
# Calculate the average math score 
Average_Math_Score = school_data_complete["math_score"].mean()
round(Average_Math_Score,6)
```

```python
# Calculate the average reading score
Average_Reading_Score = school_data_complete["reading_score"].mean()
round(Average_Reading_Score, 6)
```

```python
# Percentage_Passing_Math

# extract all math_score >= 70 from school_data_complete dataframe
School_Score_math =school_data_complete[(school_data_complete["math_score"] >= 70)]

# find out count of students with math score >= 70
mathcore_greater70=School_Score_math["Student ID"].count()

# Calculate the percentage of students with a passing math score (70 or greater) 
#i.e.(mathcore_greater70/Total_Students)*100 
Percentage_Passing_Math = (mathcore_greater70/ Total_Students) *100
round(Percentage_Passing_Math, 6)
```

```python
# Percentage_Passing_Reading  

# extract all reading_score >= 70 from score from school_data_complete dataframe
Reading_School_Score = school_data_complete[(school_data_complete["reading_score"] >= 70)] 

# find out students count with reading score >= 70
Readingcore_greater70 = Reading_School_Score["Student ID"].count()

# Calculate the percentage of students with a passing reading score (70 or greater)
Percentage_Passing_Reading = (Readingcore_greater70 / Total_Students) *100
round(Percentage_Passing_Reading, 6)
```

```python
# Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
Overall_Passing_Rate = (Average_Math_Score + Average_Reading_Score) / 2
round(Overall_Passing_Rate, 6)
```

```python
# Create a dataframe to hold the above results
District_Summary_df = pd.DataFrame({
                                "Total\nSchool":[Total_School],
                                "Total\nStudents":[Total_Students],
                                "Total\nBudget":[Total_Budget],
                                "Average\nMath\nScore":[Average_Math_Score],
                                "Average\nReading\nScore":[Average_Reading_Score],
                                "Percentage\nPassing\nMath":[Percentage_Passing_Math],
                                "Percentage\nPassing\nReading":[Percentage_Passing_Reading],
                                "Overall\nPassing\nRate":[Overall_Passing_Rate]
                                })
District_Summary_df
```

# 2.	PyCitySchools:- School Summary

##  all tasks are completed
-	Create an overview table that summarizes key metrics about each school, including: 
-	School Type
-	Total Students
-	Total School Budget
-	Per Student Budget
-	Average Math Score
-	Average Reading Score
-	% Passing Math
-	% Passing Reading
-	Overall Passing Rate (Average of the above two)


```python
school_type=school_data_df.set_index(["school_name"])["type"]
student_count=school_data_complete["school_name"].value_counts()

school_budget=school_data_complete.groupby(["school_name"]).mean()["budget"]
budget_perstudent=school_budget/student_count

avg_mathscore=school_data_complete.groupby(["school_name"]).mean()["math_score"]
avg_readscore=school_data_complete.groupby(["school_name"]).mean()["reading_score"]

passing_math=school_data_complete[(school_data_complete["math_score"] >= 70)]
passing_Read=school_data_complete[(school_data_complete["reading_score"] >= 70)] 


passing_math_count=passing_math.groupby(["school_name"]).count()["student_name"] / student_count *100
passing_Read_count=passing_Read.groupby(["school_name"]).count()["student_name"] / student_count *100

overall_pass_rate= (passing_math_count + passing_Read_count)/2

School_summary_df= pd.DataFrame({"School Type":school_type,
                                 "Total Student":student_count,
                                 "Total School Budget":school_budget,
                                 "Budget Per Student":budget_perstudent,
                                 "Average Math Score":avg_mathscore,
                                 "Average Reading Score":avg_readscore,
                                 "Percentage Passing Math":passing_math_count,
                                 "Percentage Passing Read":passing_Read_count,
                                 "Percent Overall Passing Rate":overall_pass_rate
                                })
School_summary_df
```

# # 2.4.	PyCitySchools :- Scores by School Spending 
 
 
##  all tasks are completed

•	Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. 
Include in the table each of the following: 
-	Average Math Score
-	Average Reading Score
-	% Passing Math
-	% Passing Reading
-	Overall Passing Rate (Average of the above two)


```python
student_count=school_data_complete["school_name"].value_counts()
school_budget=school_data_complete.groupby(["school_name"]).mean()["budget"]

budget_perstudent=school_budget/student_count

avg_mathscore=school_data_complete.groupby(["school_name"]).mean()["math_score"]
avg_readscore=school_data_complete.groupby(["school_name"]).mean()["reading_score"]

passing_math=school_data_complete[(school_data_complete["math_score"] >= 70)]
passing_Read=school_data_complete[(school_data_complete["reading_score"] >= 70)] 

passing_math_count=passing_math.groupby(["school_name"]).count()["student_name"] / student_count *100
passing_Read_count=passing_Read.groupby(["school_name"]).count()["student_name"] / student_count *100

overall_pass_rate= (passing_math_count + passing_Read_count)/2

SchoolSpending_summary_df= pd.DataFrame({
                                 "Budget Per Student":budget_perstudent,
                                 "Average Math Score":avg_mathscore,
                                 "Average Reading Score":avg_readscore,
                                 "Percentage Passing Math":passing_math_count,
                                 "Percentage Passing Read":passing_Read_count,
                                 "Percent Overall Passing Rate":overall_pass_rate
                                        })
SchoolSpending_summary_df
```

```python
max = School_summary_df["Budget Per Student"].max()
print(max)
mean = School_summary_df["Budget Per Student"].mean()
print(mean)
min = School_summary_df["Budget Per Student"].min()
print(min)
```

```python
# Create bins based upon per student school budget
bins=[0,585,615,645,660]
group_labels=["< 585", "585 to 615", "615 to 645", "645 t0 660"]


pd.cut(SchoolSpending_summary_df["Budget Per Student"], bins, labels=group_labels)
```

```python
SchoolSpending_summary_df["Budget Per Student Group"] = pd.cut(SchoolSpending_summary_df["Budget Per Student"], 
                                                               bins, labels=group_labels)

SchoolSpending_summary_df
```

```python
# Create a GroupBy object based upon "Budget Per Student Group"
Spending_Group = SchoolSpending_summary_df.groupby("Budget Per Student Group")

# Find how many rows fall into each bin
print(Spending_Group["Budget Per Student"].count())

# Get the average of each column within the GroupBy object
Spending_Group[["Average Math Score", "Average Reading Score", "Percentage Passing Math","Percentage Passing Read",
                "Percent Overall Passing Rate"]]
```

```python
Scores_by_School_Spending_df= SchoolSpending_summary_df.groupby("Budget Per Student Group").mean()
Scores_by_School_Spending_df
```

```python
# hide/drop "Budget Per Student" column
Budget_Per_Student_Group_df = Scores_by_School_Spending_df.copy()
for col in ["Budget Per Student"]:
    Budget_Per_Student_Group_df=Budget_Per_Student_Group_df.drop(col,axis=1)
Budget_Per_Student_Group_df
```
